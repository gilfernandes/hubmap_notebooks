```python
import numpy as np
import cv2
```

```python
IMG_PATH = '/home/jupyter/data_2/test/aa05346ff_mask.png'
```

```python
from PIL import Image

Image.MAX_IMAGE_PIXELS = None

png = np.array(Image.open(str(IMG_PATH)), dtype=np.uint8)
```

```python
png_color = cv2.cvtColor(png, cv2.COLOR_GRAY2BGR)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(15,15))
ax[0].imshow(png)
ax[1].imshow(png_color)
```

```python
# convert the image to grayscale
gray_image = cv2.cvtColor(png_color, cv2.COLOR_BGR2GRAY)
```

```python
# convert the grayscale image to binary image
ret, thresh = cv2.threshold(gray_image,127,255,0)

# find contours in the binary image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
img = cv2.drawContours(gray_image, contours, -1, (0,255,0), 3)
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(15,15))
ax.imshow(img)
```
